```python
from __future__ import print_function
from bs4 import BeautifulSoup
import requests
import time
import urllib
import re
import os
import sys
```

```python
def solveChapter(url):
    tryTimes = 3
    wb_data=''
    while(tryTimes > 0):
        try:
            wb_data = requests.get(url)
            break
        except Exception as e:
            tryTimes -= 1
            print('try ' + str(tryTimes))
            time.sleep(1)
    if tryTimes <= 0:
        sys.exit()
    wb_data.encoding = 'gb18030'
    soup = BeautifulSoup(wb_data.text, 'lxml')
    article = soup.select('#txt > p')##txt > p:nth-child(15)  
    title = soup.select('#nr_title')[0].text  ##nr_title

    #sort article
    sortedArticle=list()
    for i in range (len(article)):
        for paragraph in article:
            if int(paragraph.get("data-id"))==i :#compare the para-id and the position
                sortedArticle.append(paragraph)
                break
    writeArticle(title,sortedArticle)
```

```python
def writeArticle(title,sortedArticle):
    #write title
    f.write(title)
    f.write('\r\n')
    #segmentation and write content
    for paragraph in sortedArticle:
        text = paragraph.text
        text = text.replace('[舞文小说网www. 首发]','')
        for line in text.split('\xa0\xa0\xa0\xa0'):
            if line == '':
                continue
            f.write('\t'+line)
            f.write('\r\n')
    f.write('\r\n')
    f.flush()
```

```python
#https://wap.5wx.org/read/40/40921/
wb_data = requests.get('https://wap.5wx.org/read/40/40921/')
wb_data.encoding = 'gb18030'
soup = BeautifulSoup(wb_data.text, 'lxml')
```

```python
book = soup.select('#newlist > div')
#sort book
sortedBook=list()
for i in range (len(book)):
    for chapter in book:
        if int(chapter.get("data-id"))==i :#compare the para-id and the position
            sortedBook.append(chapter)
            break
```

```python
f = open("xiaoyi.txt","a+",encoding='utf-8') 
#get urls
#flag = False
for part in sortedBook:
    partChapters = part.select('li > a')
    for chapter in partChapters:
        #pid = chapter.get('href')
        #pid = pid.replace(".html",'')
        #if int(pid) == 83147:
        #    flag = True
        #if flag:
        url = 'https://wap.5wx.org/read/40/40921/' + chapter.get('href')
        solveChapter(url)
        print(url)
        time.sleep(0.5)
```
